```python
import numpy as np

def loan(amt, pymt, intst, mnths):
    mnthly = intst / (12 * 100)
    print("Month Amount   Interest  Remaining")
    if mnths == 'all':
        mnths = 2400
    for i in range(int(np.round(mnths))):
        i_amt = amt + (mnthly * amt)
        n_amt = i_amt - pymt
        print(f" {int(i) + 1}    {amt:.2f} {i_amt:.2f}  {n_amt:.2f}")
        amt = n_amt
        if amt <= pymt:
            amt = 0
            mnths = i + 2
            break
    return f'{amt:.2f} after {int(mnths / 12)} years and {mnths % 12} months'

loan(11000, 100, 4.36, 12)
```

```python
import numpy as np

def compounding_loan(amt, pymt, intst, days):
    dyly = intst / (365.25 * 100)
    print("Day  Amount  Interest Remaining")
    if days == 'all':
        days = 365.25 * 200
    for i in range(int(np.round(days))):
        i_amt = amt + (dyly * amt)
        n_amt = i_amt - pymt
        print(f" {int(i) + 1}   {amt:.2f}  {i_amt:.2f}  {n_amt:.2f}")
        amt = n_amt
        if amt <= pymt:
            amt = 0
            days = i + 2
            break
    print(f'{amt:.2f} after {int(days / 365.25)} years and '
         f'{int(np.round((days % 365.25) / 30.5))} months')

def nelnet_loan(prin, pymt, intst, days):
    dyly = intst / (365.25 * 100)
    print("Day  Amount  Interest Remaining")
    if days == 'all':
        days = 365.25 * 200
    amt = prin
    for i in range(int(np.round(days))):
        i_amt = amt + (dyly * prin)
        n_amt = i_amt - pymt
        print(f" {int(i) + 1}   {amt:.2f}  {i_amt:.2f}  {n_amt:.2f}")
        amt = n_amt
        if amt <= pymt:
            amt = 0
            days = i + 2
            break
    print(f'{amt:.2f} after {int(days / 365.25)} years and '
         f'{int(np.round((days % 365.25) / 30.5))} months')

nelnet_loan(7000, 0, 4.99, 100)
```
